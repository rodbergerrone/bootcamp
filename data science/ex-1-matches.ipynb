```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('matches.csv', parse_dates=['date'], low_memory=False)
```

```python
df
```

```python
df[df.country == "Italy"]
```

```python
df[df.home_team == "Juventus"]
```

```python
df[(df.date.dt.year.between(2010, 2015)) & 
   (df.home_team.isin(["Roma", "Juventus"])) &
  (df.away_team.isin(["Roma", "Juventus"]))]
```

```python
df['deference'] = abs(df.home_team_goal - df.away_team_goal)
df
```

```python
df.nlargest(10, 'deference')
```

```python
df.groupby('season').aggregate({'home_team_goal': 'sum', 'away_team_goal': 'sum'}).plot(kind='bar')
```
